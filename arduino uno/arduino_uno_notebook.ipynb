```python
import numpy as np
MAXN = 14000
st = f"const byte MAXN = {MAXN};\nconst int PROGMEM arrays[] = " + "{"
for i in range(MAXN):
    st += f"{np.random.randint(0, 256)}, "
st += "};"
print(st)
```

```python
import serial
arduino = serial.Serial(port='COM6',   baudrate=2000000, timeout=.1)
x = []
y = []
cont = 0
while cont < 12000:
    content = arduino.readline().rstrip()
    if(b", " in content and len(content) < 1000):
        data = str(content.decode()).split(", ")
        x.append(int(data[0]))
        y.append(float(data[1])/(20*int(data[0])))
        print(f"{len(x)}: ({data[0]}, {data[1]})")
        cont += 1
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[200:], [i*4 for i in y][200:], marker='o', s=2)
plt.ylabel("Microseconds")
plt.xlabel("Array size (bytes)")
plt.title("Time to increment array")
plt.show()
```

```python
st = "{"
import numpy as np
for l in range(1, 820, 820//(1 + 15)):
    entry = "{"
    arr = np.zeros(l, dtype=int)
    for i in range(l):
        arr[i] = i
    np.random.shuffle(arr)
    ind = arr[l - 1]
    final = np.zeros(l, dtype=int)
    for i in range(l):
        final[ind] = arr[i]
        ind = final[ind]
    for i in range(820):
        if i < l:
            entry += str(final[i]) + ", "
        elif i != 820 - 1:
            entry += "0, "
        else:
            "0"
    entry += "},\n"
    st += entry
st += "};"
print(st)
```

```python
import serial
arduino = serial.Serial(port='COM6',   baudrate=2000000, timeout=.1)
xl = []
yl = []
cont = 0
while cont < 17:
    content = arduino.readline().rstrip()
    if(len(content) > 0):
        print(content)
        data = str(content.decode()).split(", ")
        xl.append(int(data[0]))
        yl.append(float(data[1])/(17*int(data[0])))
        print(f"{len(xl)}: ({data[0]}, {data[1]})")
        cont += 1
```

```python
import matplotlib.pyplot as plt
plt.plot([i*2 for i in xl][1:], yl[1:], marker='o')
plt.ylabel("Microseconds")
plt.xlabel("Array size (bytes)")
plt.title("Time to traverse random array")
plt.show()
```
